```python
import librosa
import IPython
import matplotlib.pyplot as plt
from librosa import display
import numpy as np
%matplotlib inline
```

```python
y, sr = librosa.load("../data/TheTourist-171222.mp3", sr=9000)
#Shift+tab to get functions
```

```python
sr
```

```python
y.shape
```

```python
y.shape[0]/sr
```

```python
IPython.display.Audio(data=y[1000:150000], rate=sr)
```

```python
Y = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=96)
Y_log = librosa.power_to_db(Y, ref=np.max)
```

```python
plt.figure(figsize=(10, 4))
display.specshow(Y_log, y_axis='mel', x_axis='time', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```
